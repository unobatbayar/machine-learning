```python
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(10)]
y = [2*i for i in range(10)]
```

```python
print(x)
print(y)
```

```python
plt.xlabel("x - axis")
plt.ylabel("y - axis")
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```
